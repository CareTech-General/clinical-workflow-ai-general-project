```python
# ── Medical Appointment No-Shows ──────────────────────────────────────────────
# Source: https://www.kaggle.com/datasets/joniarroba/noshowappointments
# Task:   Predict whether a patient will miss their scheduled appointment
# File:   KaggleV2-May-2016.csv
# ─────────────────────────────────────────────────────────────────────────────

# TODO: import any additional libraries you need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/KaggleV2-May-2016.csv')

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df.head()
```

```python
# Dataset-specific setup — run this before anything else
# The original dataset has column name typos and a text target; fix them here

df = df.rename(columns={
    'Hipertension': 'Hypertension',
    'Handcap':      'Handicap',
    'SMS_received': 'SMS_Received',
    'No-show':      'no_show',
})

# Encode target: 'Yes' → 1 (no-showed), 'No' → 0 (showed up)
df['no_show'] = (df['no_show'] == 'Yes').astype(int)

# Parse date columns (useful for feature engineering later)
df['ScheduledDay']   = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])

# Remove the one erroneous age value (-1)
df = df[df['Age'] >= 0].copy()

print("Shape after cleanup:", df.shape)
print(f"No-show rate: {df['no_show'].mean():.1%}")
df.head()
```

```python
# Quick look — run this before anything else
print("Missing values:")
print(df.isnull().sum())

print("\nTarget distribution (no_show):")
print(df['no_show'].value_counts())

# Things to think about:
# - The date columns aren't useful raw — what features could you engineer from them?
# - Neighbourhood has 80+ unique values. How will you handle that?
```
